```python
duplicates = df.duplicated().any()
print(duplicates)
```

```python
top_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)
print(top_category)
```

```python
df['Order Date'].head(10)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Quarter'] = df['Order Date'].dt.to_period('Q')
df['Quarter Num'] = df['Order Date'].dt.quarter
df['Quarter Num'].head(10)
```

```python
quarterly_performance = df.groupby('Quarter Num')['Amount'].sum().reset_index()
```

```python
print(quarterly_performance)
```

```python
quarterly_performance = quarterly_performance.sort_values(by='Amount',ascending= False)
```
