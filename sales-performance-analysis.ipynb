```python
top_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)
print(top_category)
```

```python
df['Order Date'].head(10)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Quarter'] = df['Order Date'].dt.to_period('Q')
df['Quarter Num'] = df['Order Date'].dt.quarter
df['Quarter Num'].head(10)
```

```python
quarterly_performance = df.groupby('Quarter Num')['Amount'].sum().reset_index()
```

```python
print(quarterly_performance)
```

```python
quarterly_performance = quarterly_performance.sort_values(by='Amount',ascending= False)
```

```python
print(quarterly_performance)
```

```python
state_perf =  df.groupby('State')['Amount'].sum().sort_values(ascending= False).head(5)
```

```python
print(state_perf)
```

```python
city_perf = (
    df.groupby('City')['Amount']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(city_perf)
```

```python
payment_counts = df['PaymentMode'].value_counts()
print(payment_counts)
```

```python
payment_mode_revenues = df.groupby('PaymentMode')['Amount'].sum().reset_index()
print(payment_mode_revenues)
```

```python
highest_revenue_payment =  payment_mode_revenues.sort_values(by = 'Amount',ascending = False)
print(highest_revenue_payment)
```

```python
ord_count = df.groupby('Order ID').size().reset_index(name = 'Order Count')
print(ord_count)
```

```python
repeat_customer_df = ord_count[ord_count['Order Count'] > 1].sort_values(by='Order Count', ascending=False)
print(repeat_customer_df)
```

```python
customer_spending = df.groupby('Order ID')['Amount'].sum().reset_index()
```
